1- Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

```python
# seu código aqui...

valor = False
while valor == False:
  opcao = float(input('Digite um número entre 0 a 10: '))
  if opcao >= 0 and opcao <= 10:
    valor = True
    break
  print("Valor inválido, tente novamente")
```

3- Faça um programa que leia e valide as seguintes informações:
*   Nome: maior que 3 caracteres;
*   Idade: entre 0 e 150;
*   Salário: maior que zero;
*   Sexo: 'f' ou 'm';
*   Estado Civil: 's', 'c', 'v', 'd';

**Use a função `len(string)` para saber o tamanho de um texto (número de caracteres).**

```python
# seu código aqui...
nome = input("Digite seu nome: ")
while len(nome) <= 3:
  print("O nome deve ter mais que 3 caracteres")
  nome = input("Digite seu nome novamente: ")

idade = int(input("Digite sua idade: "))
while idade < 0 or idade > 150:
  print("A idade deve ser entre 0 e 150")
  idade = int(input("Digite sua idade novamente: "))

salario = int(input("Digite seu salário: "))
while salario <= 0:
  print("O salário deve ser maior que 0")
  salario = int(input("Digite seu salário novamente: "))

sexo = input("Digite seu sexo, \"F\" para feminino e \"M\" para masculino: ").lower()
while sexo not in ['f', 'm']:
  print("Digite um sexo válido")
  sexo = input("Digite seu sexo novamente: ").lower()

estado_civil = input("Digite seu estado civil, \"S\" para solteiro, \"C\" para casado, \"V\" para viúvo e \"D\" para divorciado: ").lower()
while estado_civil not in ['s', 'c', 'v', 'd']:
  print("Digite um estado civil válido")
  estado_civil = input("Digite seu estado civil novamente: ").lower()
```

4- Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
# seu código aqui...

populacao_a = 80000
populacao_b = 200000
ano = 0
while populacao_a < populacao_b:
  populacao_a *= 1.03
  populacao_b *= 1.015
  ano += 1
print("Serão necessários",ano,"anos para que a população do país A ultrapasse ou iguale a população do país B")
```

6- Faça um programa que leia 5 números e informe a soma e a média dos números.

```python
# seu código aqui...

numeros = []
soma = 0
media = 0

for i in range(1,6):
  numeros.append(int(input("Digite um número: ")))
for i in range(len(numeros)):
  soma += numeros[i]
media = soma / 5
print("A soma dos seus números é",soma,"e a média dos seus números é",media)
```

7- Faça um programa que receba dois números inteiros e gere os números inteiros que estão no intervalo compreendido por eles.

```python
# seu código aqui...

num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
step = int((num2 - num1) / abs(num2 - num1))
for i in range(num1+step,num2,step):
  print(i, end = " ")
```

9- Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da turma varia entre 0 e 25,26 e 60 e maior que 60; e então, dizer se a turma é jovem, adulta ou idosa, conforme a média calculada.

```python
# seu código aqui...
soma = 0
pessoas = int(input("Digite a quantidade de pessoas na turma: "))
for i in range(1,pessoas+1):
  idade = int(input("Digite a idade da "+ str(i) + "° pessoa: "))
  soma += idade
media = soma / pessoas
if media >= 0 and media <= 25:
  print("A turma é jovem")
elif media <= 60:
  print("A turma é adulta")
else:
  print("A turma é idosa")
```

10- Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
# seu código aqui...

eleitores = int(input("Digite a quantidade de eleitores: "))
voto = ""
candidato_a = 0
candidato_b = 0
candidato_c = 0
for i in range(1,eleitores+1):
  voto = input("Vote na letra do seu candidato [A/B/C]").lower()
  while voto not in ['a', 'b', 'c']:
    voto = input("Candidato inválido, tente novamente: ").lower()
  if voto == "a":
    candidato_a += 1
  elif voto == "b":
    candidato_b += 1
  else:
    candidato_c += 1
  voto = ""
print("Candidato A:",candidato_a,"votos")
print("Candidato B:",candidato_b,"votos")
print("Candidato C:",candidato_c,"votos")
```
